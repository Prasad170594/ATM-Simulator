```python
class ATM_Simulator:
    def __init__(self):
        self.balance = 10000
        self.pin = 0000

    def verify_pin(self, user_pin):
        return user_pin == self.pin

    def balance_display(self):
        print(f"Your current balance is Rs.{self.balance}")

    def withdraw(self, amt):
        if amt > self.balance:
            print("Insufficient Balance")
        else:
            self.balance = self.balance - amt
            print('Rs.',amt,' has been withdrawn. Your current balance is Rs.',self.balance)

    def deposit(self, amt):
        self.balance = self.balance + amt
        print('Rs.',amt,' has been deposited. Your cuurent balance is Rs.',self.balance)

    def exit(self):
        print("Thank you for using our services. Have a great day!")

def main():
    atm = ATM_Simulator()
    print("***** Welcome To ATM Machine Simulator *****")
    
    user_pin = int(input("Enter Your Pin: "))

    if user_pin != atm.pin:
        print("Incorrect PIN.")
        return

    while True:
        print("""
                Options you can Exercise are:
                1) Balance
                2) Withdraw
                3) Deposit
                4) Exit
        """)
        
        option = int(input("Select Your Transaction from the above options: "))
        
        if option == 1:
            atm.balance_display()
        elif option == 2:
            amt = int(input("Enter Amount: "))
            if amt <= 0:
                print("Amount must be greater than zero.")
            else:
                atm.withdraw(amt)
        elif option == 3:
            amt = int(input("Enter Amount: "))
            if amt <= 0:
                print("Amount must be greater than zero.")
            else:
                atm.deposit(amt)
        elif option == 4:
            atm.exit()
            break
        else:
            print("Invalid option. Please select a number between 1 and 4.")

main()
```
